```python
import numpy as np
import pandas as pd
```

```python
heights = pd.read_csv('data/heights.csv')
heights.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(heights['midparentHeight'], heights['childHeight'], alpha = 0.5)
plt.grid()
plt.xlabel('mid parent height (in inches)')
plt.ylabel('child height (in inches)')
plt.show()
```

```python
mpht = 66
heights.query("@mpht - 0.5 < midparentHeight < @mpht + 0.5")['childHeight'].mean()
```

```python
def avg_child_ht(mpht):
    pred_chld_ht = heights.query("@mpht - 0.5 < midparentHeight < @mpht + 0.5")['childHeight'].mean()
    return pred_chld_ht
```

```python
pred_chld_hts = heights['midparentHeight'].map(avg_child_ht)
pred_chld_hts
```

```python
plt.scatter(heights['midparentHeight'], heights['childHeight'], alpha = 0.5)
plt.scatter(heights['midparentHeight'], pred_chld_hts, c='red')
plt.grid()
plt.xlabel('mid parent height (in inches)')
plt.ylabel('child height (in inches)')
plt.show()
```
